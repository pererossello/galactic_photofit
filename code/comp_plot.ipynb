```python
names = ['fit_sersic_exp', 'fit_sersic_exp_and_sersic_bump', 'fit_sersic_exp_own_mask']
dic = {}

for i, name in enumerate(names):
    x_from_file = []
    dif_from_file = []
    with open(f"../results/{name}_dif.txt", "r") as f:
        next(f)  # skip the header
        for line in f:
            xi, di = map(float, line.strip().split("\t"))
            x_from_file.append(xi)
            dif_from_file.append(di)

    ds = xr.open_dataset(f"../results/{name}.nc")

    dic[name] = {'x': x_from_file, 
                 'dif': dif_from_file,
                 'ds': ds}
```

```python
res, rat, subplots = 1080, 1.6, (2,3)
fig, axs, fs, gs = pu.initialize_figure(ratio=None, fig_w=rat*res, fig_h=res, subplots=subplots, theme=None, hr=[2,1],
                                        hmerge=[1])


lab = '$\mu_{r,\\text{data}} - \mu_{r,\\text{fit}}$'
R = 0.396
tls=1.2
vm_res = 0.9
colors = ['k', 'r', 'b']
labels = ['Sersic+Exp', 'Sersic+Exp & Sersic (bump)', 'Sersic+Exp (new mask)']
lss = ['-', '--', '-.']
chis = [3.38, 2.50, 1.82]
AICS = [50886.79, 37641.39, 114583.56]
BICS = [50970.57 ,37778.48, 114683.14 ]
for i in range(len(names)):
    nam = names[i]
    x = np.array(dic[nam]['x'])*R
    im = pu.magplot(dic[nam]['ds']['data_res_mag'].values,
    axs[0][i], fs, vms=[-vm_res,vm_res], 
    mask=dic[nam]['ds']['mask'].values, tls=tls,
    cmap='RdBu', cbar=False, xtop=False, yright=False, title=labels[i])

    text = f'$\chi^2 = {chis[i]:.2f}$\n$\\text{{AIC}}={AICS[i]:.2e}$\n$\\text{{BIC}}={BICS[i]:.2e}$'
    box = dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3', linewidth=0.05*fs)
    axs[0][i].text(0.975, 0.975, text , transform=axs[0][i].transAxes, fontsize=1*fs, ha='right', va='top', color='k',
    bbox=box)


    axs[1][0].plot(x, dic[nam]['dif'], label=labels[i], color=colors[i], ls=lss[i], lw=0.2*fs)

    if i==2:
        ax = axs[0][i]
        ls = 1
        wdth=0.15
        lgth = 0.35
        shrink, pad = 0.7, 0.02
        cbar_label=lab
        cbar = plt.colorbar(im, ax=ax, orientation='vertical', 
                    location='right', shrink=shrink, pad=pad*fs)
        cbar.set_label(cbar_label, size=1.5*ls * fs, labelpad=0.2*fs)  # Label for the color bar
        cbar.ax.tick_params(labelsize=tls * fs, width=wdth * fs, length=lgth * fs)
        ticks = cbar.get_ticks()
        for spine in cbar.ax.spines.values():
            spine.set_linewidth(0.15 * fs) 



axs[1][0].set_ylabel(lab, fontsize=1.3*fs)
axs[1][0].set_xlabel('Distance (arcsec)', fontsize=1.2*fs)
axs[1][0].set_xlim(0, x[-1])
axs[1][0].set_ylim(-0.495, 0.425)

axs[1][0].tick_params(axis='both', which='major', labelsize=tls*fs, 
)

legend = axs[1][0].legend(loc=[0.65,0.525], handlelength=0.25*fs, handletextpad=0.2*fs,
labelspacing=0.0175*fs, frameon=False)
frame = legend.get_frame()  # Get the bounding box (legend frame)
frame.set_linewidth(0.2*fs)  # Set the linewidth of the legend box (Optional)
# legend._legend_box.sep = fs
# To change the text color in the legend
for text in legend.get_texts():
    text.set_color("black")
    text.set_fontsize(fs*1.2)



labels = ['a)', 'b)', 'c)', 'd)']
x = [0.31, 0.605, 0.895]
y = [0.91, 0.375]

# for i, x in enumerate(x_coords):
#     for j, y in enumerate(y_coords):
#         index = 3*j + i
for i in range(len(x)):
    fig.text(x[i], y[0], labels[i], fontsize=fs*1.25,
    fontweight='bold', color='darkred')

fig.text(x[2], y[1], labels[3], fontsize=fs*1.25,
fontweight='bold', color='darkred')

savefold = '../figures/'
if not os.path.exists(savefold):
    os.makedirs(savefold)
filename = f'residual_comparison.png'
savepath = os.path.join(savefold, filename)

# save figure
fig.savefig(savepath, dpi=300, bbox_inches='tight')

plt.close()
```
